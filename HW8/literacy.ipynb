```python
import matplotlib.pyplot as pl
import pandas as pd
import numpy as np
import scipy.stats as stat
pl.style.use('fivethirtyeight')
%pylab inline 
```

```python
#http://data.un.org/Default.aspx
df = pd.read_csv('literacydata.csv')
```

```python
mask = (df['Sex'] == 'All genders')
df = df[mask]
```

```python
df_cut = df[(df['Reference Area'] == 'MDG_Northern Africa') |
        (df['Reference Area'] == 'MDG_Southern Asia') | 
        (df['Reference Area'] == 'MDG_Sub-Saharan Africa') |
        (df['Reference Area'] == 'MDG_Western Asia') |
        (df['Reference Area'] == 'MDG_South-eastern Asia') |
        (df['Reference Area'] == 'MDG_Oceania') |
        (df['Reference Area'] == 'MDG_Latin America') |
        (df['Reference Area'] == 'MDG_Eastern Asia') |
        (df['Reference Area'] == 'MDG_Caucasus and Central Asia') |
        (df['Reference Area'] == 'MDG_Caribbean')]
```

```python
df_cut = df_cut.sort('Time Period')
```

```python
df_s_asia = df_cut[df_cut['Reference Area'] == 'MDG_Southern Asia']
df_w_asia = df_cut[df_cut['Reference Area'] == 'MDG_Western Asia']
df_se_asia = df_cut[df_cut['Reference Area'] == 'MDG_South-eastern Asia']
df_e_asia = df_cut[df_cut['Reference Area'] == 'MDG_Eastern Asia']
df_n_africa = df_cut[df_cut['Reference Area'] == 'MDG_Northern Africa']
df_s_africa = df_cut[df_cut['Reference Area'] == 'MDG_Sub-Saharan Africa']
df_latin = df_cut[df_cut['Reference Area'] == 'MDG_Latin America']
df_car = df_cut[df_cut['Reference Area'] == 'MDG_Caribbean']
```

```python
df_s_asia_list = df_s_asia.values.T.tolist()
df_w_asia_list = df_w_asia.values.T.tolist()
df_se_asia_list = df_se_asia.values.T.tolist()
df_e_asia_list = df_e_asia.values.T.tolist()
df_n_africa_list = df_n_africa.values.T.tolist()
df_s_africa_list = df_s_africa.values.T.tolist()
df_latin_list = df_latin.values.T.tolist()
df_car_list = df_car.values.T.tolist()
```

```python
fig = pl.figure(figsize=(12,10))

pl.plot(df_e_asia_list[1], df_e_asia_list[5], linewidth = '8', color = '#E06666', label = 'East Asia')
pl.plot(df_latin_list[1], df_latin_list[5], linewidth = '8', color = '#E69138', label = 'Latin America')
pl.plot(df_se_asia_list[1], df_se_asia_list[5], linewidth = '8', color = '#F1C232', label = 'Southeast Asia')
pl.plot(df_w_asia_list[1], df_w_asia_list[5], linewidth = '8', color = '#93C47D', label = 'West Asia')
pl.plot(df_n_africa_list[1], df_n_africa_list[5], linewidth = '8', color = '#C27BA0', label = 'North Africa')
pl.plot(df_s_asia_list[1], df_s_asia_list[5], linewidth = '8', color = '#8E7CC3', label = 'South Asia')
pl.plot(df_s_africa_list[1], df_s_africa_list[5], linewidth = '8', color = '#6FA8DC', label = 'Sub-Saharan Africa')

pl.legend(bbox_to_anchor=(1.23, 0.635))

pl.xlabel('Year', fontsize = 16)
pl.ylabel('Percent Literacy', fontsize = 16)
pl.title('Literacy by region over time', fontsize = 18)
```
