```python
import os
from scipy.optimize import curve_fit, minimize
import pylab as pl
import numpy as np
%pylab inline

md = {}
md['chisq'] = {}
md['pois'] = {}
md['gaus'] = {}
md['binomial'] = {}
md['gamma'] = {}
md['lognormal'] = {}

means = {}
means['chisq'] = {}
means['pois'] = {}
means['gaus'] = {}
means['binomial'] = {}
means['gamma'] = {}
means['lognormal'] = {}

mymean = 100
df = 100

mysize = 2000 / (np.array(range(1, 100) + [10]))

for n in mysize:
    md['chisq'][n] = np.random.chisquare(df, size = n)
    md['pois'][n] = np.random.poisson(lam = 100, size = n)
    md['gaus'][n] = np.random.normal(loc = 100, scale = 1, size = n)
    md['binomial'][n] = np.random.binomial(n = 200, p = 0.5, size = n)
    md['gamma'][n] = np.random.gamma(shape = 100, size = n)
    md['lognormal'][n] = np.random.lognormal(mean = 4, sigma = 1, size = n)
```

# 2.

```python
fig_mu_N = pl.figure(figsize=(20,15))

# Chisquare Distribution
axchisq_mu_n = fig_mu_N.add_subplot(331)
for nn in md['chisq'].iterkeys():
    means['chisq'][nn] = md['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.set_title('Chisquare')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

# Poisson Distribution
axpois_mu_n = fig_mu_N.add_subplot(332)
for nn in md['pois'].iterkeys():
    means['pois'][nn] = md['pois'][nn].mean()
    axpois_mu_n.plot(nn, means['pois'][nn], 'o')
    axpois_mu_n.set_xlabel('sample size')
    axpois_mu_n.set_ylabel('sample mean')
    axpois_mu_n.set_title('Poisson')
    axpois_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

# Gaussian Distribution
axgaus_mu_n = fig_mu_N.add_subplot(333)
for nn in md['gaus'].iterkeys():
    means['gaus'][nn] = md['gaus'][nn].mean()
    axgaus_mu_n.plot(nn, means['gaus'][nn], 'o')
    axgaus_mu_n.set_xlabel('sample size')
    axgaus_mu_n.set_ylabel('sample mean')
    axgaus_mu_n.set_title('Gaussian')
    axgaus_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

# Binomial Distribution
axbinomial_mu_n = fig_mu_N.add_subplot(334)
for nn in md['binomial'].iterkeys():
    means['binomial'][nn] = md['binomial'][nn].mean()
    axbinomial_mu_n.plot(nn, means['binomial'][nn], 'o')
    axbinomial_mu_n.set_xlabel('sample size')
    axbinomial_mu_n.set_ylabel('sample mean')
    axbinomial_mu_n.set_title('Binomial')
    axbinomial_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

# Gamma Distribution
axgamma_mu_n = fig_mu_N.add_subplot(335)
for nn in md['gamma'].iterkeys():
    means['gamma'][nn] = md['gamma'][nn].mean()
    axgamma_mu_n.plot(nn, means['gamma'][nn], 'o')
    axgamma_mu_n.set_xlabel('sample size')
    axgamma_mu_n.set_ylabel('sample mean')
    axgamma_mu_n.set_title('Gamma')
    axgamma_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')

# Lognormal Distribution
axlognormal_mu_n = fig_mu_N.add_subplot(336)
for nn in md['lognormal'].iterkeys():
    means['lognormal'][nn] = md['lognormal'][nn].mean()
    axlognormal_mu_n.plot(nn, means['lognormal'][nn], 'o')
    axlognormal_mu_n.set_xlabel('sample size')
    axlognormal_mu_n.set_ylabel('sample mean')
    axlognormal_mu_n.set_title('Log Normal')
    axlognormal_mu_n.plot([min(mysize), max(mysize)],[df,df], 'k')
```

The plots do look as expected. As the sample size increases, the sample means of the distributions become closer to the set distribution means. This is because the variability of the distributions decreases as sample size increases.

```python
# Plot all sample means
allmeans = []
for n in means['chisq']:
    allmeans.append(means['chisq'][n])
    allmeans.append(means['pois'][n])
    allmeans.append(means['gaus'][n])
    allmeans.append(means['binomial'][n])
    allmeans.append(means['gamma'][n])
    allmeans.append(means['lognormal'][n])

pl.figure(figsize = (10, 10))
pl.hist(allmeans, bins = 30)
pl.xlabel('sample mean')
pl.ylabel('N')
pl.title('Sample Mean For All Distributions')
```
